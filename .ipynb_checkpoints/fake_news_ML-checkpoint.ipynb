```python
combine_df = pd.concat((pd.read_csv(file).assign(Label=file) 
                       for file in pop_files),ignore_index=True)
combine_df = combine_df.replace("data/Fake.csv","fake")
combine_df = combine_df.replace("data/True.csv","true")
combine_df
```

# Tokenization

```python
def tokeniz(text):
    token = re.split('\W+', text)
    return token

combine_df["token_text"] = combine_df["text"].apply(lambda x:tokeniz(x.lower()))

combine_df.head()
```

# Remove stopwords

```python
nltk.download('stopwords')
```

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopword(token):
    text = [word for word in token if word not in stopword]
    return text

combine_df["nostopword_text"] = combine_df["token_text"].apply(lambda x:remove_stopword(x))

combine_df.head()
```

# Lemmatizing

```python
# nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wl = WordNetLemmatizer()
```

```python
def lemmatizing(stopword):
    text = [wl.lemmatize(word) for word in stopword]
    return text

combine_df["lemmatizing_text"] = combine_df["nostopword_text"].apply(lambda x:lemmatizing(x))

combine_df.head()
```

```python
combine_df.to_csv("data/clean_data.csv")
```
